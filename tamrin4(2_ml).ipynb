```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import numpy as np
from PIL import Image
import cv2

from dataset import load_hoda
```

```python
X_train, y_train, X_test, y_test = load_hoda(1000,200,15)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
image = Image.open("persian-digits.jpg")
plt.imshow(image)
```

```python
image = image.convert('L')
img = np.array(image)
img2 = img[212:288,180:220]
#plt.imshow(img2,cmap='gray')

img2 = cv2.resize(img2,(15,15))

# Apply thresholding to create a binary image
_, img2 = cv2.threshold(img2, 128, 255, cv2.THRESH_BINARY)
img2 = 255-img2

plt.imshow(img2,cmap='gray')
img2 = np.reshape(img2, [-1,15**2])

prediction = knn.predict(img2)
print("Prediction:", prediction)
```
